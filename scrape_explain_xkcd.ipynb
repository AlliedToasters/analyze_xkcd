```python
import scrapy
import re
from scrapy.crawler import CrawlerProcess
```

```python

urls = iter(['http://www.explainxkcd.com/wiki/index.php/{}'.format(x) for x in range(1, 1942)])
start = next(urls)


class XKCDSpider(scrapy.Spider):
    name = "xkcd"
    
    start_urls = [
        start,
    ]

    def parse(self, response):
            
            
        yield {
            'title': response.css('tr+ tr b').extract(),
            'title_text': response.css('br+ span').extract(),
            'transcript': response.css('#mw-content-text > dl dd').extract(),
        }
        next_page = 'http://www.explainxkcd.com/wiki/index.php/1941'
        
        try:
            next_page = response.urljoin(next(urls))
            yield scrapy.Request(next_page, callback=self.parse)
        except:
            pass

process = CrawlerProcess({
    'FEED_FORMAT': 'json', 
    'FEED_URI': 'data.json', 
    'LOG_ENABLED': False, 
    'ROBOTSTXT_OBEY': True,
    'USER_AGENT': 'ThinkfulDataScienceBootcampCrawler (thinkful.com)',
    'AUTOTHROTTLE_ENABLED': True,
    'HTTPCACHE_ENABLED': True
})

process.crawl(XKCDSpider)
process.start()
print('Success!')
```
